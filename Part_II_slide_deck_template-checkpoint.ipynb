## Effect of  borrower interest rate from Prosper Loan Data


## Investigation Overview
Prosper is  loan company, The purpose of this project is to look at the primary factors of loans and borrowers interest rate that might be used to predict their progression. 
## Dataset Overview

> The Prosper loan dataset comprises of 113937 loan entries with 81 attributes on each loan which include so many attribute of borrowers eg employement status , occupation, creditgrade etc

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe

loan_data=pd.read_csv('prosperLoanData.csv')
```

```python
#selectig varibles needed for analysis
loan=loan_data[['ProsperRating (Alpha)',
                        'Term',
                        'LoanStatus',
                        'BorrowerAPR',
                        'BorrowerRate',
                        'ListingCreationDate',
                        'BorrowerState',
                        'Occupation',
                        'EmploymentStatus',
                        'LoanOriginalAmount',
                        'DebtToIncomeRatio',
                        'StatedMonthlyIncome',
                        'MonthlyLoanPayment',
                        'ProsperScore'
               ]]
```

```python
# convert string to ordinal category type
ordinal_rating = ['AA', 'A', 'B', 'C', 'D', 'E','HR']
ordered_var = pd.api.types.CategoricalDtype(ordered = True, categories = ordinal_rating)
loan['ProsperRating'] = loan['ProsperRating (Alpha)'].astype(ordered_var)
```

## Loan Status 
> Over 80% of all loans on Prosper are either current or paid completed. 10.5% of Prosper's loans were charged off and 4.4% all went into default. Additionally, 1.8% of debts were lumped  under the classification of Past DueÂ and 0,2% went into finalpaymentprogress

```python
# Aggregate  loans 
loan['LoanStatus'] = loan['LoanStatus'].apply(lambda x: x if 'Past Due' not in x else 'Past Due')

# Convert LoanStatus to a categorical variable
status = ['Defaulted','Chargedoff', 'Past Due', 'Current', 'FinalPaymentInProgress', 'Completed']

loan_status = pd.api.types.CategoricalDtype(ordered = True, categories = status)

loan['LoanStatus'] = loan['LoanStatus'].astype(loan_status)
```

```python
# loan status distribution

# sum of the loan status
loans=loan['LoanStatus'].value_counts().sum()
# count value and order the counts
status_count=loan['LoanStatus'].value_counts()
status_order=status_count.index


base_color = sb.color_palette()[0]
sb.countplot(data=loan, x='LoanStatus', color=base_color, order=status_order);

# get the current tick locations and labels
locs, labels = plt.xticks(rotation=45) 

# loop through each pair of locations and labels
for loc, label in zip(locs, labels):
    # get the text property for the label to get the correct count
    count = status_count[label.get_text()]
    pct_string = '{:0.1f}%'.format(100*count/loans)

    # print the annotation just below the top of the bar
    plt.text(loc, count+2, pct_string, ha = 'center', color = 'black')
    
plt.title('Loan Status Distribution');
```

## Loan Original Amount Distribution

> There are three peaks at 4K, 10K, and 15K, as can be seen. These three loan sizes appear to be the most popular ones that people have taken out.

```python
# Plot the loan original amount against a log scale because it has a long tail

log_binsize = 0.025
bins_log = 10 ** np.arange(3, np.log10(loan['LoanOriginalAmount'].max())+log_binsize, log_binsize)
plt.hist(data = loan, x = 'LoanOriginalAmount', bins = bins_log)
plt.xscale('log')
x_ticks = [1000, 2000, 3000, 4000, 5000, 10000, 15000,20000, 30000]
x_ticks_names = ['1k','2k', '3k', '4k', '5k', '10k','15k', '20k','30k']
plt.xticks(x_ticks, x_ticks_names)
plt.xlim((1000,35000))
plt.xlabel('Loan Original Amount ($)')
plt.title('Loan Original Amount Distribution');
```

## Relation between Borrower Rate and Prosper Score, Prosper Rating
There is compelling evidence that the borrower's rate is very well predicted by the borrower's Prosper rating or score. We can infer that the rate is lower the higher the score. In a similar vein, borrowers with AA prosper ratings should anticipate an APR that will probably fall between 2-15%, while those with poor ratings (HR) can anticipate interest rates as high as 40%!



```python
 # ploting a violin bar to show relationship between borrower rate ,proper score and proper rating
def violin(var, color):
    fig, ax = plt.subplots(ncols = 2, figsize = [10,6])
    sb.violinplot(data = loan, y = var, x = 'ProsperScore', ax = ax[0],
                  color = color)
    ax[0].set_xlabel("Prosper Score")
    ax[0].set_ylabel(var.capitalize())
    sb.violinplot(data = loan, y = var, x = 'ProsperRating', ax = ax[1],
               color = color)

    ax[1].set_xlabel("Prosper Rating")
    ax[1].set_ylabel("")
    fig.suptitle('Relation between Borrower Rate and Prosper Score, Prosper Rating')
    plt.show()
    
violin('BorrowerRate', sb.color_palette()[0])
```

## Borrower Rate by Prosper Rating and Loan terms
We note that the rates applied on 5-year loans are generally higher than the rates applied on other maturities, regardless of the prosper rating (with the exception of the E rating (same rates for 3 and 5 years loans ) and the HR rating)

```python
#plot Borrower Rate by Prosper Rating and Loan terms
fig = plt.figure(figsize = [8,5])
ax = sb.pointplot(data = loan, x = 'ProsperRating', y = 'BorrowerRate', hue = 'Term',
           palette = 'Blues', linestyles = '',dodge = 0.4, ci='sd')
plt.title('Borrower Rate by Prosper Rating and Loan terms');
```
